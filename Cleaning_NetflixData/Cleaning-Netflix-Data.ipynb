```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/netflixdata.csv")

#1 INSPECTING THE DataFrame with.info .describe and .head
# df.info()
# df.describe()
# df.head()

#2 IDENTIFING MISSING VALUES
# df.isnull().sum()

#3 HANDLING MISSING VALUES BY COUMN
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Not Available")
df["country"] = df["country"].fillna("Unknown")
df = df.dropna(subset=["date_added"])
df["rating"] = df["rating"].fillna(df["rating"].mode()[0])
#df.isnull().sum()

#df["duration"].sample(10)

#4 FIX MIX-TYPE columns (e.g., duration stored as "90 min")
df["duration_num"] = df["duration"].str.extract(r'(\d+)')
df["duration_type"] = df["duration"].str.extract(r'([A-Za-z]+)')
df = df.dropna(subset=["duration_num"])
df["duration_num"] = df["duration_num"].astype(int)

#df.info()



df["date_added"] = pd.to_datetime(
    df["date_added"],
    format="%B %d, %Y",
    errors="coerce"
)
# df["date_added"].dtype

df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month
df["day_added"] = df["date_added"].dt.day

#df.info()
df.to_csv("data/netflixdata_cleaned.csv", index=False)
```
